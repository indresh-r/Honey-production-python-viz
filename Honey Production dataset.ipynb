# Data Visualization on Honey Production dataset using seaborn and matplotlib libraries.

<span style="color:blue">**Domain**</span>: **Food and agriculture**

<span style="color:blue">**Context:**</span>
**In 2006, a global concern was raised over the rapid decline in the honeybee population, an integral component
to American honey agriculture. Large numbers of hives were lost to “Colony-Collapse-Disorder”, a
phenomenon of disappearing “worker-bees” causing the remaining “hive-colony” to collapse. Speculation
around the cause of this disorder points to hive-diseases and pesticides harming the pollinators, though no
overall consensus has been reached. Twelve years later, some industries are observing recovery but the
American honey industry is still largely struggling. The U.S. used to locally produce over half the honey it
consumes per year. Now, honey mostly comes from overseas, with 350 of the 400 million pounds of honey
consumed every year originating from imports. This dataset provides insight into honey production supply and
demand in America by state from 1998 to 2012.**

<span style="color:blue">**Objective:**</span>
**The Goal is to use Python visualization libraries such as seaborn and matplotlib to investigate the data and get
some useful conclusions**

**3. Display the percentage distribution of the data in each year using the pie chart.**


```python
plt.figure(figsize=(10,6))
df['year'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('Distribution of the data in each year')
plt.show()
```

**4. Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings.**


```python
plt.figure(figsize=(12, 8))
sns.displot(data=df, x='priceperlb', kind='hist', kde=True)
plt.title('Distribution of price per lb')
plt.show()
print('skewness of priceperlb is: ',df['priceperlb'].skew())
```

<span style="color:blue">**INFERENCE:**</span>

<span style="font-family: Arial; font-size: 16px;">**◉ The analysis of the 'price per lb' column reveals a right-skewed distribution, signifying that the majority of honey prices per pound tend to cluster towards the lower end of the scale. The presence of a longer right tail suggests occasional instances of higher-priced honey, potentially indicative of premium products or unique market dynamics.This may due to presence of outliers.**</span>




**5. Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and
write your findings.**


```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x= 'numcol',y= "prodvalue")
plt.title("Relationship between the numcol and prodvalue")
plt.xlabel('numcol')
plt.ylabel('prodvalue')
plt.show()
```

<span style="color:blue">**INFERENCE:**</span>

<span style="font-family: Arial; font-size: 16px;">**◉ The scatterplot depicting the relationship between the number of honey-producing colonies ('numcol') and production value ('prodvalue') reveals a clear and positive linear trend. As the number of colonies increases, there is a corresponding rise in production value.**</span>




**6. Plot and understand the relationship between categorical variable 'year' and a numerical variable
'prodvalue' through boxplot, and write your findings.**


```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='year', y='prodvalue')
plt.title('Relationship between Year and Production Value')
plt.xlabel('Year')
plt.ylabel('Production Value (prodvalue)')
plt.show()
```

<span style="color:blue">**INFERENCE:**</span>
   
<span style="font-family: Arial; font-size: 16px;">**◉ The median production value has increased over the years, indicating a positive trajectory for honey production.The consistent box widths suggest similar variability across different years.**</span>

<span style="font-family: Arial; font-size: 16px;">**◉ While there is some variation, the overall trend suggests a positive relationship between year and production value.This aligns with the recovery observed in the honey industry after the Colony Collapse Disorder crisis.**</span>
